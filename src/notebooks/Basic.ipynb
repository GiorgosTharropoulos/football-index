```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
argentina_file_path = './Sample Data/raw stats/Argentina-Primera Division.xlsx'
england_file_path = './Sample Data/raw stats/England-Premier League.xlsx'
france_file_path = './Sample Data/raw stats/France-Ligue 1.xlsx'


argentina = pd.read_excel(argentina_file_path, index_col='ID')
england = pd.read_excel(england_file_path, index_col='ID')
france = pd.read_excel(france_file_path, index_col='ID')
```

```python
argentina.dropna(axis='columns', how='all', inplace=True)
argentina.describe()
```

```python
dfs = [argentina, england, france]
all_nan_cols = []
for df in dfs:
    for col in df.columns:
        if len(df[col]) == len(df[col][df[col].isna()]):
            all_nan_cols.append(df[col].name)
            
            
print(all_nan_cols)
```

```python
from collections import Counter
Counter(all_nan_cols)
```

```python
for df in dfs:
    df.dropna(inplace=True, how='all', axis='columns')
```
